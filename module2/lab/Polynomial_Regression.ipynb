# Polynomial Regression


Estimated time needed: **40** minutes

What if your data is more complex than a straight line? Surprisingly, you can use a linear model to fit nonlinear data. A simple way to do this is to add powers of each feature as new features, then train a linear model on this extended set of features. This technique is called Polynomial Regression.

There are two factors when determining model performance: overfitting and underfitting. Overfitting is when the model is too complex and does well on the training data but not on the test data. Underfitting is when the model is too simple and performs poorly on the training and testing data sets. 

Overfitting is simple to deal with, using methods like regularization, which we will discuss in the next lab. To deal with underfitting, we can build a more complex model using methods like polynomial regression. If making a more complex model does not work, this may involve using more data to train the model on or obtaining new features. As this process is complex, it's better to determine if the model can overfit the data first. Therefore, in this section, we will use Polynomial Regression to overfit the data to determine if we have an adequate amount of data.

In this notebook, we will explore Polynomial Regression and perform polynomial transform using individual features as well as multiple features.

# Mathematical Formulas and Statistical Methods Reference

This notebook demonstrates polynomial regression and advanced machine learning techniques. Below are the mathematical foundations with properly formatted equations:

## 1. Polynomial Regression Fundamentals

### Simple Linear Regression
$$Y = \beta_0 + \beta_1 X + \varepsilon$$

### Polynomial Regression of Degree p
$$Y = \beta_0 + \beta_1 X + \beta_2 X^2 + \beta_3 X^3 + \cdots + \beta_p X^p + \varepsilon$$

**Components:**
- $Y$: Target variable (dependent variable)
- $X$: Original feature variable  
- $X^2, X^3, \ldots, X^p$: Polynomial features
- $\beta_i$: Regression coefficients
- $p$: Polynomial degree
- $\varepsilon$: Random error term

### Multiple Features with Polynomial Transform
For two features $X_1, X_2$ with degree 2:

$$Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \beta_3 X_1^2 + \beta_4 X_1 X_2 + \beta_5 X_2^2 + \varepsilon$$

**Feature expansion includes:**
- Original: $X_1, X_2$
- Squared: $X_1^2, X_2^2$  
- Interaction: $X_1 X_2$

## 2. Bias-Variance Tradeoff

### Expected Error Decomposition
$$\mathbb{E}[\text{Error}] = \text{Bias}^2 + \text{Variance} + \text{Irreducible Error}$$

Where:

$$\text{Bias}[\hat{f}(x)] = \mathbb{E}[\hat{f}(x)] - f(x)$$

$$\text{Variance}[\hat{f}(x)] = \mathbb{E}[(\hat{f}(x) - \mathbb{E}[\hat{f}(x)])^2]$$

**Model Complexity Trade-offs:**
- **Underfitting:** High Bias, Low Variance
- **Optimal:** Balanced Bias and Variance  
- **Overfitting:** Low Bias, High Variance

## 3. Evaluation Metrics

### R-squared (Coefficient of Determination)
$$R^2 = 1 - \frac{\text{SS}_{\text{res}}}{\text{SS}_{\text{tot}}}$$

Where:
$$\text{SS}_{\text{res}} = \sum_{i=1}^n (y_i - \hat{y}_i)^2$$
$$\text{SS}_{\text{tot}} = \sum_{i=1}^n (y_i - \bar{y})^2$$
$$\bar{y} = \frac{1}{n}\sum_{i=1}^n y_i$$

**Range:** $R^2 \in [0, 1]$ where 1 = perfect fit

### Mean Squared Error
$$\text{MSE} = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2$$

### Root Mean Squared Error  
$$\text{RMSE} = \sqrt{\text{MSE}} = \sqrt{\frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2}$$

## 4. Data Preprocessing

### StandardScaler (Z-score Normalization)
$$x_{\text{std}} = \frac{x - \mu}{\sigma}$$

Where:
$$\mu = \frac{1}{n}\sum_{i=1}^n x_i$$
$$\sigma = \sqrt{\frac{1}{n-1}\sum_{i=1}^n (x_i - \mu)^2}$$

**Result:** $\mathbb{E}[x_{\text{std}}] = 0$, $\text{Var}[x_{\text{std}}] = 1$

### MinMaxScaler 
$$x_{\text{scaled}} = \frac{x - x_{\min}}{x_{\max} - x_{\min}}$$

**Result:** $x_{\text{scaled}} \in [0, 1]$

## 5. Feature Selection

### F-statistic for Regression
$$F = \frac{\text{MS}_{\text{regression}}}{\text{MS}_{\text{residual}}} = \frac{\frac{\text{SS}_{\text{regression}}}{p}}{\frac{\text{SS}_{\text{residual}}}{n-p-1}}$$

Where:
- $p$: Number of features
- $n$: Number of samples

$$\text{SS}_{\text{regression}} = \sum_{i=1}^n (\hat{y}_i - \bar{y})^2$$
$$\text{SS}_{\text{residual}} = \sum_{i=1}^n (y_i - \hat{y}_i)^2$$

**Higher F-score = More predictive feature**

## 6. Cross-Validation

### K-Fold Cross-Validation Error
$$\text{CV Error} = \frac{1}{K} \sum_{k=1}^{K} \text{Error}_k$$

### Standard Error of CV
$$\text{SE}_{\text{CV}} = \sqrt{\frac{1}{K-1} \sum_{k=1}^{K} (\text{Error}_k - \overline{\text{Error}})^2}$$

Where $\overline{\text{Error}}$ is the mean CV error.

## 7. Grid Search Optimization

### Objective Function
$$\theta^* = \arg\min_{\theta \in \Theta} \frac{1}{K} \sum_{k=1}^{K} L(f_{\theta}, \text{Data}_k)$$

Where:
- $\theta$: Hyperparameter vector
- $\Theta$: Parameter search space (grid)
- $L$: Loss function
- $f_{\theta}$: Model with parameters $\theta$

## 8. Pipeline Mathematics

### Function Composition
$$\hat{y} = f_3(f_2(f_1(X)))$$

Where for polynomial pipeline:
- $f_1$: StandardScaler: $f_1(X) = \frac{X - \mu}{\sigma}$
- $f_2$: PolynomialFeatures: $f_2(X) = [X, X^2, \ldots, X^p]$
- $f_3$: LinearRegression: $f_3(X) = X\beta$


## 9. Train-Test Split

### Dataset Partition
For dataset $\mathcal{D}$ with $n$ samples:
$$\mathcal{D} = \mathcal{D}_{\text{train}} \cup \mathcal{D}_{\text{test}}$$
$$|\mathcal{D}_{\text{train}}| = (1-\alpha) \cdot n$$
$$|\mathcal{D}_{\text{test}}| = \alpha \cdot n$$

Where $\alpha$ = test_size (commonly 0.2 or 0.3)

## 10. Model Performance Analysis

### Performance Gap Metric
$$\text{Generalization Gap} = R^2_{\text{train}} - R^2_{\text{test}}$$

**Interpretation:**
- Gap $< 0.05$: Excellent generalization
- Gap $0.05 - 0.15$: Good generalization
- Gap $0.15 - 0.25$: Moderate overfitting
- Gap $> 0.25$: Severe overfitting

## 11. Statistical Distributions

### Normal Distribution PDF
$$f(x; \mu, \sigma^2) = \frac{1}{\sigma\sqrt{2\pi}} \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$$

**Properties:**
- $\mathbb{E}[X] = \mu$
- $\text{Var}[X] = \sigma^2$
- $\int_{-\infty}^{\infty} f(x)dx = 1$

## 12. Advanced Concepts

### Feature Interactions
For interaction effects between $X_1$ and $X_2$:
$$\frac{\partial^2 Y}{\partial X_1 \partial X_2} = \beta_{\text{interaction}}$$

### Regularization (Preview)

**Ridge Regression (L2):**
$$\min_{\beta} \left\{ \frac{1}{2n}\|y - X\beta\|_2^2 + \lambda\|\beta\|_2^2 \right\}$$

**Lasso Regression (L1):**
$$\min_{\beta} \left\{ \frac{1}{2n}\|y - X\beta\|_2^2 + \lambda\|\beta\|_1 \right\}$$

### Ordinary Least Squares Solution
$$\hat{\beta} = (X^T X)^{-1} X^T y$$

**Normal Equation:**
$$X^T X \beta = X^T y$$

## Mathematical Summary

**Key Insight:** Polynomial regression transforms non-linear relationships into linear algebra problems by creating polynomial features. The optimal degree $d^*$ minimizes:

$$d^* = \arg\min_d \left\{ \text{Bias}^2(d) + \text{Variance}(d) \right\}$$

where:
- $\text{Bias}(d) \downarrow$ as degree increases
- $\text{Variance}(d) \uparrow$ as degree increases

**Always validate with cross-validation to find the optimal balance!**


# Mathematical Formulas Used in Polynomial Regression

## 1. Polynomial Regression
For degree $d$:
$$Y = \beta_0 + \beta_1 X + \beta_2 X^2 + \ldots + \beta_d X^d + \varepsilon$$

Where:
- $Y$: Target variable
- $X$: Feature variable
- $\beta_i$: Coefficients learned by least squares
- $\varepsilon$: Error term

## 2. Mean Squared Error (MSE)
$$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

Where:
- $y_i$: Actual value
- $\hat{y}_i$: Predicted value
- $n$: Number of samples

## 3. R-squared (Coefficient of Determination)
$$R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}$$

Where:
- $\bar{y}$: Mean of actual values
- Range: $[0, 1]$ where 1 indicates perfect fit

## 4. Standardization (Z-score)
$$x' = \frac{x - \mu}{\sigma}$$

Where:
- $x$: Original value
- $\mu$: Mean of feature
- $\sigma$: Standard deviation of feature
- Result: Mean = 0, Standard deviation = 1

## 5. Pipeline Concept
A pipeline chains multiple preprocessing steps:
$$\text{Raw Data} \xrightarrow{\text{StandardScaler}} \text{Scaled Data} \xrightarrow{\text{PolynomialFeatures}} \text{Poly Data} \xrightarrow{\text{LinearRegression}} \text{Predictions}$$

Mathematical representation:
$$\hat{y} = f_3(f_2(f_1(X)))$$

Where:
- $f_1$: StandardScaler transformation
- $f_2$: PolynomialFeatures transformation
- $f_3$: LinearRegression prediction

## 6. Train-Test Split
Data partitioning:
$$\mathcal{D} = \mathcal{D}_{\text{train}} \cup \mathcal{D}_{\text{test}}$$

Where:
- $|\mathcal{D}_{\text{train}}| = (1-\alpha) \times n$
- $|\mathcal{D}_{\text{test}}| = \alpha \times n$
- $\alpha$: test_size ratio (commonly 0.2 or 0.3)
- $n$: total number of samples

**Variables:**
- $X_{\text{train}}, X_{\text{test}}$: Feature matrices
- $y_{\text{train}}, y_{\text{test}}$: Target vectors

## 7. Hyperparameter Grid Search
Finds the best combination of parameters using cross-validation:
$$\theta^* = \arg\min_{\theta \in \Theta} \frac{1}{K} \sum_{k=1}^{K} \text{CV\_Error}_k(\theta)$$

Where:
- $\theta$: Hyperparameter vector
- $\Theta$: Parameter search space (grid)
- $K$: Number of cross-validation folds
- $\text{CV\_Error}_k(\theta)$: Error on fold $k$ with parameters $\theta$

**Common Grid Search Parameters:**
- Polynomial degree: $d \in \{1, 2, 3, 4, 5\}$
- Regularization parameter: $\lambda \in \{0.01, 0.1, 1.0, 10.0\}$

## 8. Cross-Validation Error
K-Fold Cross-Validation:
$$\text{CV Error} = \frac{1}{K} \sum_{k=1}^{K} \text{Error}_k$$

Standard Error:
$$\text{SE}_{\text{CV}} = \sqrt{\frac{1}{K-1} \sum_{k=1}^{K} (\text{Error}_k - \overline{\text{Error}})^2}$$

## 9. Polynomial Feature Generation
Number of polynomial features for $n$ variables and degree $d$:
$$\text{Number of features} = \binom{n + d}{d} = \frac{(n+d)!}{d! \cdot n!}$$

**Examples:**
- 1 feature, degree 3: $\binom{1+3}{3} = 4$ features: $[1, X, X^2, X^3]$
- 2 features, degree 2: $\binom{2+2}{2} = 6$ features: $[1, X_1, X_2, X_1^2, X_1X_2, X_2^2]$

## 10. Model Selection Criteria
Performance gap analysis:
$$\text{Overfitting Gap} = R^2_{\text{train}} - R^2_{\text{test}}$$

**Interpretation:**
- Gap $< 0.05$: Good generalization
- Gap $0.05 - 0.15$: Acceptable
- Gap $> 0.15$: Overfitting detected

## GridSearch and  Pipeline


In this section of the notebook, we will define a pipeline object, then use GridSearch to find the best hyper-parameters of the model by using  cross-validation method of the  parameter grid, as shown in the figure below. We create `PolynomialFeatures()` pipeline.


````python
Input=[ ('scaler', StandardScaler()), ('polynomial', PolynomialFeatures(include_bias=False,degree=2)), ('model',LinearRegression())]
pipe=Pipeline(Input)
````

To search for the best combination of hyperparameters, we create a  <b>GridSearchCV</b> object with dictionary of parameter values. The parameters of pipelines can be set using the name of the key separated by 
'__', then the parameter name of the key.


````python
param_grid = {
    "polynomial__degree": [1, 2, 3]
    
    
}
````

<b>polynomial__degree</b>: is the degree of the polynomial. 




````python
search = GridSearchCV(pipe, param_grid, n_jobs=1)
````

````python
pipe.fit(X_train, y_train)
````

````python
search.fit(X_test, y_test)
````

The object finds the best parameter values on the validation data. We can obtain the estimator with the best parameters and assign it to the variable ```best```, as follows:


````python
best=search.best_estimator_
best
````

We see the degree is one and normalize is <code>Ture </code>.


Now, we test our model on the test data to see the best score.


````python
best.score(X_test,y_test)
````

Finally, we can plot a distribution of the predicted values versus the actual values. 


````python
predicted=best.predict(X_test)
plot_dis(y_test,predicted)
````

As we see, the result is almost perfect!
